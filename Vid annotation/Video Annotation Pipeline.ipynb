```python
def pipelineD(dataFile, startingTime, Participant):
    date_format = "%H:%M:%S.%f"

    startingTime = datetime.strptime(startingTime,date_format)
    
    with open(dataFile) as f:
        lines = f.readlines()
    temp = []
    for k in lines:
        temp.append(k.split(' ', 2))
    mat = pd.DataFrame(temp)
    currentMat=mat.drop(mat.columns[0], axis=1)
    currentMat.columns = ['time', 'annotation']
    currentMat.dropna(subset=['time'], inplace=True)
    currentMat['time'] = pd.to_datetime(currentMat['time'], format="%H:%M:%S.%f")
    currentMat['time'] = currentMat['time'] - startingTime
    currentMat.to_csv('c:/Users/Viraj/Internship/Motion Artifacts FNIRS/Vid annotation/timestamps participants/D/Participant_' + Participant + '_D.csv', index=False, mode = 'w')    
    currentMat = pd.read_csv('c:/Users/Viraj/Internship/Motion Artifacts FNIRS/Vid annotation/timestamps participants/D/Participant_' + Participant + '_D.csv')
    currentMat['time'] = currentMat['time'].str[7:]
    currentMat.to_csv('c:/Users/Viraj/Internship/Motion Artifacts FNIRS/Vid annotation/timestamps participants/D/Participant_' + Participant + '_D.csv', index=False, mode = 'w')    

    return currentMat
```
